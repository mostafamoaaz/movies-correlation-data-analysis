```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (12,8) # configure the figure size for visualization purpose
```

```python
df = pd.read_csv(r'movies.csv')
df.head()
```

```python
#checking if there are nulls aka missing values

# we could have populate the missing values of specific columns with the mean of the rest values
# but we droped these values because it contain non numerical values and it only occupy 0.28% of the data

for col in df.columns:
    missing_percentage = np.mean(df[col].isnull())
    print("{}-{:.2f}%".format(col,missing_percentage))
    
df.dropna(how='any', inplace=True)
```

```python
df.dtypes

```

```python
df["votes"] = df["votes"].astype("int64")
df["budget"] = df["budget"].astype("int64")
df["gross"] = df["gross"].astype("int64")
df["runtime"] = df["runtime"].astype("int64")
df.head()
```

```python
#the column "year" and "release date" are not the same 
#so we extract the year portion of release date and create another column called "year_corrected"

import re
count = 0
for ind in df.index:
    x = re.search("(\d{4})",df['released'][ind])
    if str(df['year'][ind]) != x.group():
        count+=1
print("the number of incorrect data in \'year\' column are : "+str(count))

#(\d{4})
```

```python
df['year_corrected'] = df['released'].str.extract(r'(\d{4})')
count = 0
for ind in df.index:
    x = re.search("(\d{4})",df['released'][ind])
    if  x.group() != df['year_corrected'][ind]:
        count+=1
print("the number of incorrect data in \'year\' column are : "+str(count))
```

```python
#order the DF by gross value descendingly
df.sort_values(by=["gross"], inplace=False, ascending=False)
```

```python
#count duplicates and drop any duplicate row

print("number of duplicated data is : " +str(df.duplicated( keep='first').sum()))

print("number of rows before dropping : "+ str(df.shape[0]))

df.drop_duplicates()

print("number of rows after dropping : "+ str(df.shape[0]))
```

```python
#correlation matrix 
# create a heatmap to see the correlation between numeric values e.g(gross,budget,score,votes)
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("correlation matrix")
```

```python
# create scatter plot of (budget,gross)

plt.scatter(x = df["budget"], y = df["gross"])

plt.title("budget vs gross")

plt.xlabel("Budget")

plt.ylabel("Gross revenue")

plt.show
```

```python
# create scatter plot of (budget,gross) using seaborn to see the the correlation line

sns.regplot(x = "budget",y = "gross", data = df,scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# create scatter plot of (votes,gross)


sns.regplot(x = "votes",y = "gross", data = df,scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```
